```python
q1 = np.quantile(df['Suscribers'], 0.25)

q3 = np.quantile(df['Suscribers'], 0.75)

med = np.median(df['Suscribers'])
 
iqr = q3-q1
 
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(iqr, upper_bound, lower_bound)
```

```python
outliers = df['Suscribers'][(df['Suscribers'] <= lower_bound) | (df['Suscribers'] >= upper_bound)]
outliers
```

```python
avg_subscribers = df['Suscribers'].mean()
avg_likes = df['Likes'].mean()
avg_comments = df['Comments'].mean()

above_avg_streamers = df[
    (df['Suscribers'] > avg_subscribers) &
    (df['Likes'] > avg_likes) &
    (df['Comments'] > avg_comments)
]
above_avg_streamers
```

```python
top_10_streamers = above_avg_streamers.nlargest(5, 'Likes')
top_10_streamers = above_avg_streamers.nlargest(5, 'Comments')
top_10_streamers = above_avg_streamers.nlargest(5, 'Suscribers')
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(18, 6))

axes[0].bar(top_10_streamers['Username'], top_10_streamers['Likes'])
axes[0].set_ylabel('Likes')
axes[0].set_title('Likes of Above Average Streamers')

axes[1].bar(top_10_streamers['Username'], top_10_streamers['Comments'])
axes[1].set_ylabel('Comments')
axes[1].set_title('Comments of Above Average Streamers')

axes[2].bar(top_10_streamers['Username'], top_10_streamers['Suscribers'])
axes[2].set_ylabel('Subscribers')
axes[2].set_title('Subscribers of Above Average Streamers')

plt.tight_layout()

plt.show()
```
